```python
x,y,z,xd,yd,zd,xdd,ydd,zdd,mu,r1,r2,gam = symbols('x,y,z,xdot,ydot,zdot,xddot,yddot,zddot,mu,r_1,r_2,gamma')
```

```python
r1e = sqrt((x+mu)**2+y**2 + z**2)
r2e = sqrt((x - (1-mu))**2+y**2 + z**2)
r1e,r2e
```

```python
V = -((1-mu)/r1e + mu/r2e)
V
```

```python
xdde = 2*yd +x - diff(V,x).subs(r1e,r1).subs(r2e,r2)
xdde
```

```python
ydde = -2*xd + y - diff(V,y).subs(r1e,r1).subs(r2e,r2)
ydde
```

```python
zdde = - diff(V,z).subs(r1e,r1).subs(r2e,r2)
zdde
```

moving $x$ to $x - (1 - \mu + \gamma)$

```python
feh = expand(xdde.subs(r2,r2e).subs(r1,r1e).subs(x,x - (1 - mu + gam)))
```

```python
simplify(feh.subs(x**2,0).subs(y**2,0).subs(z**2,0))
```
